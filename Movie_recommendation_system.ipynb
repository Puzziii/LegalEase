```python
import numpy as mp
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('movies_metadata.csv', encoding='ISO-8859-1', on_bad_lines='skip', engine='python')
```

```python
df = df.drop_duplicates().reset_index(drop = True)
```

```python
df.duplicated().sum()
```

```python
df = df[['title', 'overview', 'genres', 'tagline', 'vote_average', 'popularity']]
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
```

```python
df.isnull().sum()
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
df.isnull().sum()
```

```python
df.iloc[0]['genres']
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x : " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df.head()
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
df.isnull().sum()
```

```python
df['tags'] = df['overview'] + ' ' + df['genres'] + ' ' + df['tagline']
```

```python
df.head()
```

```python
df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  text = str(text).lower() # convert to lowercase
  text = re.sub(r'[^\w\s]', '', text) # Remove punctuations using regex
  words = text.split() # Split each words
  words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words] # remove stop words and tokenise each word
  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
df.head()
```

```python
df['tags'][1]
```

```python
indices = pd.Series(df.index, index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000, ngram_range=(1, 2), stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n = 10): # n = 10 is 10 movie reccomendation
  if title not in indices:
    return ['Movie Not Found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1] # Get Top 10 recomendation
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```

```python
import pickle

pickle.dump(tfidf_matrix, open('tfidf_matrix.pkl', 'wb'))
pickle.dump(indices, open('indices.pkl', 'wb'))

df.to_pickle('df.pkl')

pickle.dump(tfidf, open('tfidf.pkl', 'wb'))
```
